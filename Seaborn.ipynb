<h1>Seaborn</h1>

Seaborn offers a variety of functionality which makes it useful and easier than other frameworks.

<h2>Functionalities</h2>
<ul>
<li>A function to plot statistical time series data with flexible estimation and representation of uncertainty around the estimate.</li>
<li>Functions for visualizing univariate and bivariate distributions or for comparing them between subsets of data.</li>
<li>Functions that visualize matrices of data and use clustering algorithms to discover structure in those matrices.</li>
<li>High-level abstractions for structuring grids of plots that let you easily build complex visualizations.</li>
<li>Several built-in themes for styling matplotlib graphics.</li>
<li>Tools for choosing color palettes to make beautiful plots that reveal patterns in your data.</li>
<li>Tools that fit and visualize linear regression models for different kinds of independent and dependent variables</li></ul>

<h1>Difference between Matplotlib and Seaborn</h1>

<ul>
<li>Matplotlib is highly customizable, but it can be complicated at the same time as it is hard to know what settings to tweak to achieve a good looking plot.</li>
<li>Seaborn comes with a number of themes and a high-level interface for controlling the look of matplotlib figures.</li>
<li>Example of sinwave plot in matplotlib and seaborn. We can easily tweak the sytle of seaborn.</li></ul>



```python
#matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(sum(map(ord, "aesthetics")))

#Define a simple plot function, to plot offset sine waves
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
sinplot()
plt.show()
```
